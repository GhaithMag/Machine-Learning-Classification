# III) Features Importances ou Choix des variables pour entrainer un modèle

<p>
<li>Avec la méthode de chi2, nous allons séléctioner les variables les plus importanes pour entrainer notre modèle. </li>
</p>

<li> Pour ceux qui ne sont pas famillier avec chi2 voici un <a href=" https://towardsdatascience.com/chi-square-test-for-feature-selection-in-machine-learning-206b1f0b8223
">article </a> sympa qui explique très bien ce concept et son application en ML. </li>


```python
from sklearn.feature_selection import chi2
X = data.drop('benign_0__mal_1',axis=1)
y = data['benign_0__mal_1']
chi_scores = chi2(X,y)

#On va créer un graphe avec les pvalues de nos variables
dset = pd.DataFrame()
dset['attr'] = X.columns
dset['importance'] = chi_scores[1]
dset = dset.sort_values(by='importance', ascending=False)


plt.figure(figsize=(16, 14))
plt.barh(y=dset['attr'], width=dset['importance'], color='#1976D2')
plt.title('RFECV - Feature Importances', fontsize=20, fontweight='bold', pad=20)
plt.xlabel('Importance', fontsize=14, labelpad=20)
plt.show()
```

Si la pvaleur est supérieur à 0.5 alors la variable est indépendante et peut ne pas être utilisée pour entrainer un modèle.

On ne va garder que les variables qui ont une pvaleur inférieur à 0.5.

```python
#On créé une liste avec les varaibles qui ont une pvaleur inférieur à 0.5

c = list(dset[dset['importance']< 0.5]['attr'].values)

X = data[c]
y = data['benign_0__mal_1'].values
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=101)
```

# IV) Création du modèle

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
```

-  On normalise nos données: En effet, il y a beaucoup d'écart entre les valeurs de nos colonnes. Pour rappel la normalisation permet de transformer une donnée afin de la rendre comparable à d'autres.
<li> Plus d'information sur la normalisation  <a href="https://fr.softoban.com/normalization-data-python">ici</a> </li>


```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = DecisionTreeClassifier(min_samples_split=10,min_samples_leaf=2)
model.fit(X_train,y_train)
```

## V) Evalutation du modèle

```python
#Score Entrainement
print("Le score de train est de: ", model.score(X_train,y_train))
#Score Test
print("Le score de test est de: ", model.score(X_test,y_test))
```

Il y a un léger overfitting. Le modèle est plus performant sur les données d'entrainement que sur les données test.

- Rappel:
    - On commence à overfitter quand le score du train est supérieur au score du test
    - On commence à underfitter quand le score du train et du test est faible

```python
#On va créer une matrice de confusion

from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(model, X_test, y_test) 
```

On lit la matrice de confusion comme suit:
  
  - Sur les 72 patients malades, 68 furent détéctés tandis que 4 ne furent pas détéctés.

```python
from sklearn.metrics import classification_report,confusion_matrix

predictions = model.predict(X_test)
print(classification_report(y_test,predictions))
```

```python
accuracy_score(predictions,y_test)
```

- L'accuracy c'est la somme des vrais positifs et vrais négatifs divisée par le nombre totale de prédiction. Ici on a 93.85% de bonne prédiction. Plus sa valeur est proche de 1 plus notre modèle est bon. Avec un score de 0.93 nous avons un très bon modèle. 

- Le recall c'est le nombre de vrai positif divisé par la somme des Vrai positif  et faux négatif.  Le "recall" correspond au taux d’individus positifs détectés par le modèle. Parmis les patients prédit "Benign" combien ont été correctement prédit? D'après notre raport 93% des patients prédit begign ont été prédit correctement. Notre modèle prèdit parfaitement les personnes qui n'ont pas le cancer.

- La précision c'est le  nombre de vrai positif sur la somme des vrais positifs et faux positifs. Elle correspond au taux de prédictions correctes parmi les prédictions positives.  Parmis les patients à qui l'ont ne prédit pas de cancer   combien  sont vraiment sans risque? Ici, 91% des patients prédits "Begign" le sont réélement. 

- Qu'elle est la meilleur metrics à utiliser? Cela dépend vraiment de la situation.

# VIII) Autres modèles

-Voyons si nous ne pouvons pas avoir un modèle avec un meilleur résultat que le précedant.

### A) Changeons les paramètres de notre arbre de décision et voyons lequel est le meilleur.


```python
split =  [2,5,10,20,30,40,45]
train_score = []
test_score = []
for i in split:
    model = DecisionTreeClassifier(min_samples_split=i,min_samples_leaf=2)
    model.fit(X_train,y_train)
    #Score Entrainement
    print("Le score de train est de: ", model.score(X_train,y_train))
    train_score.append(model.score(X_train,y_train))
    #Score Test
    print("Le score de test est de: ", model.score(X_test,y_test))
    test_score.append(model.score(X_test,y_test))
```

```python
plt.plot(split, train_score, '-o', label='Train')
plt.plot(split, test_score, '-o', label='Test')
plt.legend()
plt.show()
```

- Dans l'ensemble nous obtenons de bon score (Accuracy). On obtient le meilleur score quand min_split_sample vaut 10. De plus l'overfitting reste très négligeable.

### B) RandomForestClassifiers

```python
depth =  [2,5,10,20,30,40,45]
train_score = []
test_score = []
for i in depth:
    model = RandomForestClassifier(max_depth=i, random_state=0)
    model.fit(X_train,y_train)
    #Score Entrainement
    print("Le score de train est de: ", model.score(X_train,y_train))
    train_score.append(model.score(X_train,y_train))
    #Score Test
    print("Le score de test est de: ", model.score(X_test,y_test))
    test_score.append(model.score(X_test,y_test))
```

```python
plt.plot(split, train_score, '-o', label='Train')
plt.plot(split, test_score, '-o', label='Test')
plt.legend()
plt.show()
```

Ici on peut garder un depth de valeur 2. On a un très bon score sur les valeurs test et on n'overfit pas. 
Même si les autres résultats sont excellents.
